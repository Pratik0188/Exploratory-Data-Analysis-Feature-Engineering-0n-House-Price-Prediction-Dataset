<span style="font-size:25px;">
Missing Values<br>
</span>

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

if features_with_na:
    for feature in features_with_na:
        print(feature, np.round(dataset[feature].isnull().mean(),4), '% missing values')
else:
    print('No features with missing values')
```

```python
<span style= font-size:25px;> Since there are many missing values, we need to find relationship between missing values and Sales price </span>
</br>
</br>
<span style= font-size:18px;>We'll plot a diagram for this relationship.</span>
</br>
</br>
```

```python
for feature in features_with_na:
    data= dataset.copy()
    
    # We create a variable which indicates 1 if observation was missing and 0 if it is present.
    data[feature] = np.where(dataset[feature].isnull(),1,0)
    
    # We will calculate mean SalesPrice where the information is missing
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
</br>
<span style= font-size:18px;>Here with the relation between missing values and dependent variables is clearly visible. So we need to replace this NAN values with something meaningful which we will do in feature engineering section </span>
</br>
```

```python
<span style= font-size:25px;> Numerical Variables </span>
```

```python
# list of numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != '0']
print('Number of numerical feature', len(numerical_features))

# Visualizing Numerical Variables
dataset[numerical_features].head()
```

```python
</br>
<span style= font-size:25px;> Temporal variables </span>
</br>
</br>
<span style= font-size:18px;> 
From the dataset we have 4 year variables. We have extract information from the date-time variables like number of days or years. One example in this specific scenario can be difference between the year the house was built and the year house was sold. We will perform this analysis this in feature-engineering section.    
</br>
</span>
```

```python
# List of variables that contain year-information.

year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# The content of these year variables
for feature in year_feature:
    print(feature,":", dataset[feature].unique())
```

```python
# Analyzing the temporal date-time variables.
# Checking the relation between Year House is Sold and SalePrice

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
# Comparing the difference between all years with saleprice.

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        # We will capture the differnce between year variable and the year the house was sold.
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
<span style="font-size:25px;"> Variable types</span></br>
<span style="font-size:20px;"> 
1. Continuous Variable </br>
2. Discrete Variable </br>
</span>
</br>
<span style="font-size:25px;">
Discrete Variable</span></br>
```

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count {}". format(len(discrete_feature)))
```

```python
</br>
<span style="font-size:25px;">Relationship between them and sales-price</span></br>
</br>
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
</br>
<span style="font-size:22px;"> There is a relationship between variable number and sales-price</span></br>
</br>
<span style="font-size:22px;">Continuous Variable</span></br>
```

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature]
print("Continuous feature count {}".format(len(continuous_feature)))
```

```python
# We analyze continuous values by creating histograms the distribution of continuous values
for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel("Count")
    plt.title(feature)
    plt.show()
```

```python
</br>
<span style= font-size:25px;>logarithmic Trasnformation</span>
</br>
```

```python
# Now, Using Logarithmic Transformation.

for feature in continuous_feature:
    data=dataset.copy()
    if pd.api.types.is_numeric_dtype(data[feature]):
        if 0 in data[feature].unique():
            pass
        else:
            data[feature]=np.log(data[feature])
            data['SalePrice']=np.log(data['SalePrice'])
            plt.scatter(data[feature],data['SalePrice'])
            plt.xlabel(feature)
            plt.ylabel('SalesPrice')
            plt.title(feature)
            plt.show()
```

```python
</br>
<span style= font-size:25px;>Outliers</span>
</br>
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if pd.api.types.is_numeric_dtype(data[feature]):
        if 0 in data[feature].unique():
            pass
        else:
            data[feature]=np.log(data[feature])
            data.boxplot(column=feature)
            plt.ylabel(feature)
            plt.title(feature)
            plt.show()
```
